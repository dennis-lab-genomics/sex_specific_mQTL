# Testing inflation of p values
This is a cis-QTL analysis so some inflation is expected due to the enrichment of QTL near a feature of interest (like a CpG site in the case of mQTL, or a TSS in the case of eQTL). However, there should be no systematicinflation at lower ranges of p-values.

```r
library(data.table)
library(tidyverse)
library(fastqq)
```

```r
mqtl < -fread("../../data/delahaye_richs_marginal_mecs_maf01.txt.gz")
```

```r
qq(mqtl[is.finite(p)& !is.na(p)]$p,ylim=c(0,350),xlim=c(0,5.5))
```

```r
rm(mqtl)

mqtl_sex < -fread("../../data/delahaye_richs_sex_interaction_mecs_maf01.txt.gz")
```

```r
qq(mqtl_sex[is.finite(p)& !is.na(p)]$p,ylim=c(0,350),xlim=c(0,6))
```
